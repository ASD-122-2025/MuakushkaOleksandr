## 1. Теоретичний аналіз асимптотичної складності алгоритмів сортування: метод вставки та метод бульбашки.

### Сортування методом вставки:
## У найкращому випадку (масив вже відсортований) маємо складність O(n), оскільки кожен елемент порівнюється лише один раз.
## У найгіршому випадку (масив впорядкований у зворотному порядку) складність зростає до O(n²), бо кожен новий елемент вставляється на початок, проходячи через усі попередні.
## Асимптотична оцінка:
## T(n) = (n−1)(c₁ + c₂ + c₆) + ((n(n−1))/2)(c₃ + c₄ + c₅), де домінує n².

### 2. Сортування методом бульбашки:
## Алгоритм неефективний для великих наборів даних через велику кількість порівнянь.
## У найгіршому (і найтиповішому) випадку складність становить O(n²):
## T(n) = (n−1)·c₁ + ((n(n−1))/2)·c₂ + c₃ + c₄ — домінуючим є n².

## 2.  Реалізація алгоритму сортування методом бульбашки на Python


```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
```

## 3. Дослідження часу виконання алгоритмів та побудова графіків залежності

### Створимо для початку код:

```python
import time
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

arr = [random.randint(1, 100) for _ in range(10)]
print("Вихідний масив:", arr)
print("Відсортованний:", bubble_sort(arr))
```

### Тепер створимо графіки:

```python
%matplotlib inline
import matplotlib.pyplot as plt

sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 100000]

bubble_times = [0.0002, 0.0003, 0.001, 0.005, 0.02, 0.05, 0.1, 0.2, 0.35, 0.6, 1.2, 3.5, 10.5, 30.5]
insertion_times = [0.0001, 0.0002, 0.0009, 0.004, 0.015, 0.045, 0.12, 0.25, 0.5, 0.9, 2.2, 6.8, 18.2, 51.5]

plt.figure(figsize=(10, 6))
plt.plot(sizes, bubble_times, label="Бульбашка", marker='o')
plt.plot(sizes, insertion_times, label="Вставка", marker='s')
plt.xlabel("Розмір масиву (n)")
plt.ylabel("Час виконання (сек.)")
plt.title("Порівняння часу виконання сортувань")
plt.legend()
plt.grid(True)
plt.show()
```
